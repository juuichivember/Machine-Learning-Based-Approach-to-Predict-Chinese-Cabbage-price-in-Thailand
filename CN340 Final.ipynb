<br>
<h4>ต่อไปนี้จะทำการจัดข้อมูลตามวิธีการจะกล่าวไปข้างต้น<h4>

```python
year_change = {'2551': '2008','2552': '2009','2553': '2010','2554': '2011','2555': '2012','2556': '2013',
               '2557': '2014','2558': '2015','2559': '2016','2560': '2017','2561': '2018','2562': '2019',
               '2563': '2020','2564': '2021','2565': '2022','2566': '2023','2567': '2024','2568': '2025',
}
month_abbr = {
    'ม.ค.': '01',
    'ก.พ.': '02',
    'มี.ค.': '03',
    'เม.ย.': '04',
    'พ.ค.': '05',
    'มิ.ย.': '06',
    'ก.ค.': '07',
    'ส.ค.': '08',
    'ก.ย.': '09',
    'ต.ค.': '10',
    'พ.ย.': '11',
    'ธ.ค.': '12'
}

def change_date_format(old_format):
    day, month, year = old_format.split(' ')
    for key, value in year_change.items():
        if year == key:
            year = value
    for key, value in month_abbr.items():
        if month == key:
            month = value
    if len(day) != 2:
        day = '0' + day
    return f'{year}-{month}-{day}'
    
def reformat_raw_data(dataframe):
    df = dataframe.copy()
    df['date'] = df['date'].apply(change_date_format)
    df['date'] = df['date'].apply(pd.to_datetime)
    
    split_values = df['min-max'].str.split(' - ', expand=True)
    df['min_price'] = split_values[0].astype(float)
    df['max_price'] = split_values[1].astype(float)

    df.drop(columns=['min-max'], inplace=True)
    
    if (df.iloc[-1].isnull().any()) | (df['average_price'].iloc[-1] == " "):
        df['average_price'].iloc[-1] = df['average_price'].iloc[-2]
    df['average_price'] = pd.to_numeric(df['average_price'], errors='coerce')
    return df
    
def fill_nan(df):
    df_cp = df.copy()
    
    df.set_index("date", inplace=True)
    date_range = pd.date_range(start='2021-01-04', end='2025-03-27', freq='D')
    df = df.reindex(date_range).reset_index().rename(columns={'index': 'date'})
    df['average_price'] = df['date'].map(df_cp.set_index('date')['average_price'])
    df['min_price'] = df['date'].map(df_cp.set_index('date')['min_price'])
    df['max_price'] = df['date'].map(df_cp.set_index('date')['max_price'])
    df = df.bfill()
    return df
```

```python
# Date Reformat
kana = reformat_raw_data(kana)
pakbung = reformat_raw_data(pakbung)
kuangthung = reformat_raw_data(kuangthung)

# Fill Missing dates
kana = fill_nan(kana)
pakbung = fill_nan(pakbung)
kuangthung = fill_nan(kuangthung)
```

<br>
1. ฤดูร้อน เริ่มต้นประมาณกลางเดือนกุมภาพันธ์ถึงกลางเดือนพฤษภาคม (hot) 
2. ฤดูฝน เริ่มต้นประมาณกลางเดือนพฤษภาคมถึงกลางเดือนตุลาคม (wet)
3. ฤดูหนาว เริ่มต้นประมาณกลางเดือนตุลาคมถึงกลางเดือนกุมภาพันธ์ (cold)

```python
def assign_season_from_date(date):
    month = date.month
    day = date.day

    if (month == 2 and day >= 15) or 3 <= month <= 4 or (month == 5 and day <= 15):
        return 'hot'
    elif (month == 5 and day > 15) or 6 <= month <= 9 or (month == 10 and day <= 15):
        return 'wet'
    else:
        return 'cold'

merged_df.loc[:,'season'] = merged_df.loc[:,'date'].apply(assign_season_from_date)
merged_df.head()
```

```python
# จัดการ outliers โดยการ assign ค่า outliers เป็น lower_bound หรือ upper_bound
def manage_outliers(df, column_name, k=1.5):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + k * IQR
    lower_bound = Q1 - k * IQR

    df[column_name] = df[column_name].apply(
        lambda x: upper_bound if x > upper_bound else x
    )
    
    df[column_name] = df[column_name].apply(
        lambda x: lower_bound if x < lower_bound else x
    )
    return df
    
outlier_cols = ['avg_rain', 'avg_temp', 'avg_humid', 'avg_price_kuangthung',
                'min_price_kuangthung', 'max_price_kuangthung', 'avg_price_pakbung',
                'avg_price_kana']

for col in outlier_cols:
    encoded_df = manage_outliers(encoded_df.copy(), col)
encoded_df
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.api import OLS

def calculate_vif(df, features):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = features
    vif_data["VIF"] = [variance_inflation_factor(df[features].values, i) for i in range(len(features))]
    return vif_data
    
independent_features = list(no_scale_df.columns)
independent_features.remove('avg_price_kuangthung')
independent_features.remove('date')
vif_result = calculate_vif(no_scale_df, independent_features)
print(vif_result)
```

```python
def train_test_split(df):
    n = len(df)
    train_size, val_size, test_size = 0.7, 0.15, 0.15
    train_end = int(n * train_size)
    val_end = train_end + int(n * val_size)

    train_df = df.iloc[:train_end].copy()
    val_df = df.iloc[train_end:val_end].copy()
    test_df = df.iloc[val_end:].copy()
    
    return train_df, val_df, test_df
    
X_col = ['avg_rain', 'avg_temp', 'avg_humid', 'min_price_kuangthung', 'max_price_kuangthung', 'avg_price_pakbung',
         'avg_price_kana', 'gdp', 'export', 'import', 'trade_balance', 'dayofweek', 'month', 'day', 'year', 
         'is_holiday_encoded', 'is_weekend_encoded', 'season_cold', 'season_hot', 'season_wet']
y_col = ['avg_price_kuangthung']

# no_scale_df
no_scale_df.set_index(no_scale_df['date'], inplace=True)
train_no_scale, val_no_scale, test_no_scale = train_test_split(no_scale_df)

X_train_no_scale, y_train_no_scale = train_no_scale[X_col], train_no_scale[y_col]
X_val_no_scale, y_val_no_scale = val_no_scale[X_col], val_no_scale[y_col]
X_test_no_scale, y_test_no_scale = test_no_scale[X_col], test_no_scale[y_col]


# scale_std_df
scale_std_df.set_index(scale_std_df['date'], inplace=True)
train_standard, val_standard, test_standard = train_test_split(scale_std_df)

X_train_standard, y_train_standard = train_standard[X_col], train_standard[y_col]
X_val_standard, y_val_standard = val_standard[X_col], val_standard[y_col]
X_test_standard, y_test_standard = test_standard[X_col], test_standard[y_col]


# min_minmax_df
min_minmax_df.set_index(min_minmax_df['date'], inplace=True)
train_minmax, val_minmax, test_minmax = train_test_split(min_minmax_df)

X_train_minmax, y_train_minmax = train_minmax[X_col], train_minmax[y_col]
X_val_minmax, y_val_minmax = val_minmax[X_col], val_minmax[y_col]
X_test_minmax, y_test_minmax = test_minmax[X_col], test_minmax[y_col]


# สร้าง Dictionary เพื่อเก็บชุดฟีเจอร์
feature_sets = {
    'minmax': (X_train_minmax, X_val_minmax, X_test_minmax),
    'standard': (X_train_standard, X_val_standard, X_test_standard),
    'none': (X_train_no_scale, X_val_no_scale, X_test_no_scale)
}

print(f"Training set size: {X_train_no_scale.shape}")
print(f"Validation set size: {X_val_no_scale.shape}")
print(f"Test set size: {X_test_no_scale.shape}")
```

Select Linear Regression, Random Forest, Neural Network, Support Vector Machine, and Gradient boosting to train. Then, performing cross-validation to select the best combination of model and feature set.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR

models = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(random_state=999),
    'NeuralNetwork': MLPRegressor(random_state=999, max_iter=1000),
    'SVM': SVR(),
    'GradientBoosting': GradientBoostingRegressor(random_state=999)
}
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score, KFold
import numpy as np


data_sets = {
    'minmax': {'features': X_train_minmax, 'target': y_train_minmax},
    'standard': {'features': X_train_standard, 'target': y_train_standard},
    'none': {'features': X_train_no_scale, 'target': y_train_no_scale},
}


results_df = pd.DataFrame(columns=['data_set', 'model', 'mean_cv_score', 'std_cv_score'])

for set_name, data_info in data_sets.items():
    X = data_info['features']
    y = data_info['target'].values.ravel()

    for model_name, model in models.items():
        # Pipeline: Feature Selection + Model
        pipeline = Pipeline([
            ('feature_selection', SelectKBest(score_func=f_regression, k=min(5, X.shape[1]))),
            ('regressor', model)
        ])

        # Cross-Validation
        cv = KFold(n_splits=5, shuffle=True, random_state=42)
        scores = cross_val_score(pipeline, X, y, cv=cv, scoring='neg_mean_squared_error')

        mean_score = np.mean(scores)
        std_score = np.std(scores)

        pipeline.fit(X, y)

        # เก็บผลลัพธ์ใน DataFrame
        results_df = pd.concat([results_df, pd.DataFrame({
            'data_set': [set_name],
            'model': [model_name],
            'mean_cv_score': [mean_score],
            'std_cv_score': [std_score]
        })], ignore_index=True)

display(results_df)
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

param_grids = {
    'LinearRegression': {},
    'RandomForest': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]},
    'NeuralNetwork': {'hidden_layer_sizes': [(50,), (100,), (50, 50)], 'alpha': [0.0001, 0.001]},
    'SVM': {'C': [0.1, 1, 10], 'gamma': ['scale', 'auto']},
    'GradientBoosting': {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7], 'learning_rate': [0.01, 0.1]}
}

best_models = {}
best_scores = {}

for model_name, model in models.items():
    print(f"Tuning {model_name}...")
    y_train = y_train_standard.values.ravel()
    
    param_grid = param_grids.get(model_name, {})
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(X_train_standard, y_train)

    best_models[model_name] = grid_search.best_estimator_
    best_scores[model_name] = -grid_search.best_score_
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best MSE on Validation set for {model_name}: {-grid_search.best_score_}")
```
